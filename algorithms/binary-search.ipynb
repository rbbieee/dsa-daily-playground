## What is Binary Search?

Binary search is a searching algorithm that works on sorted data

Instead of checking elements one by one, it repeatedly divides the search space into two halves

Because of this approach, binary search is much faster than linear search for large datasets

## Important Requirement

Binary search only works if the data is sorted

If the data is not sorted, the result will not be correct

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

```python
data = [1, 2, 4, 7, 9]
target = 7

result = binary_search(data, target)
result
```

## Result

The function returns index 3, which means the target value is found at the fourth position in the list

This shows that the binary search implementation works correctly on sorted data

## Time and Space Complexity

- Time Complexity: O(log n)
- Space Complexity: O(1)
