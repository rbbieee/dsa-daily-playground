# Linear Search

This notebook is part of my daily practice for learning DSA (data structure & algorithm)

for this file will focus on basic searching algorithm it is linear search

```python
def linear_search(arr, what_we_looking_for):
    for i in range(len(arr)):
        if arr[i] == what_we_looking_for:
            return i
    return -1
```

```python
data = [4, 2, 7, 1, 9]
what_we_looking_for = 7

result = linear_search(data, what_we_looking_for)
result
```

Note:
The value is the index of the element, python uses zero-based indexing, so index 2 means the element is the third item in the list

## Result

The function returns index 2, which means the target value is found at the third position in the list

This confirms that the linear search implementation works as expected

## Time and Space Complexity

- Time Complexity: O(n)
- Space Complexity: O(1)
