## What is an Array?

An array is a linear data structure that stores elements in a sequential manner

In python, arrays are commonly represented using lists

```python
arr = [10, 20, 30, 40, 50]
arr
```

## Accessing Elements

Each element in an array can be accessed using its index, python uses zero-based indexing

```python
arr[0], arr[2], arr[-1]
```

## Traversing an Array

Traversal means accessing each element in the array one by one

```python
for value in arr:
    print(value)
```

## Insertion

Elements can be added to an array using append or insert

```python
arr.append(60)
arr.insert(2, 25)
arr
```

## Deletion

Elements can be removed from an array using pop or remove

```python
arr.pop()
arr.remove(25)
arr
```

## Time Complexity Summary

- Access by index: O(1)
- Traversal: O(n)
- Insertion:
  - At the end: O(1)
  - At a specific position: O(n)
- Deletion:
  - At the end: O(1)
  - At a specific position: O(n)
