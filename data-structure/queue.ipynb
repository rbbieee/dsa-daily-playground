## What is a Queue?

A queue is a linear data structure that follows the First In, First Out (FIFO) principle

The first element added to the queue will be the first one to be removed

## Simple Example

A queue is similar to a line of people

The person who comes first will be served first

```python
queue = []
queue
```

## Enqueue Operation

Enqueue means adding an element to the back of the queue

```python
queue.append(10)
queue.append(15)
queue.append(18)
queue
```

## Dequeue Operation

Dequeue means removing an element from the front of the queue

```python
removed = queue.pop(0)
removed, queue
```

## Note

Using a Python list for a queue is simple, but removing elements from the front is not very efficient

This can be improved using collections.deque

## Time Complexity

- Enqueue: O(1)
- Dequeue (using list): O(n)
