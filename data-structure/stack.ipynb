## What is a Stack?

A stack is a linear data structure that follows the Last In, First Out (LIFO) principle

The last element added to the stack will be the first one to be removed

## Simple Example

A stack can be compared to a stack of plates

You add a plate on top and remove the top plate first

```python
stack = []
stack
```

## Push Operation

Push means adding an element to the top of the stack

```python
stack.append(10)
stack.append(20)
stack.append(30)
stack
```

## Pop Operation

Pop means removing the top element from the stack

```python
stack.pop()
stack
```

## Peek Operation

Peek is used to see the top element of the stack without removing it

```python
stack[-1]
```

## Checking if the Stack is Empty

Before popping an element, it is safer to check whether the stack is empty

```python
if not stack:
    print("Stack is empty")
else:
    print("Stack is not empty")
```

## Time Complexity

- Push: O(1)
- Pop: O(1)
- Peek: O(1)
